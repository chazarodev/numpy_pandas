```python
import pandas as pd
import numpy as np
```

```python
dictionary = {'Col1': [1,2,3,np.nan],
 'Col2': [ 4,np.nan,6,7],
 'Col3': ['a','b','c',None]}
```

```python
#Creamos un dataframe basado en el diccionario arriba creado
df = pd.DataFrame(dictionary)
```

```python
df
```

Manejo de valores nulos

```python
#isnull() nos muestra valores booleanos para indicarnos valores nulos
df.isnull()
```

```python
#Si deseamos manejar valores numericos lo realizamos de la siguiente manera donde los ceros nos indicaran que existen datos, mientras que, los 1 que son valores nulos
df.isnull() * 1
```

```python
#Podemos cambiar su valor por alguna palabra que eligamos
df.fillna('Mising')
```

```python
#Para valores númericos podemos optar por llenar los datos con operaciones como la media
df.fillna(df.mean())
```

```python
#Para valores numericos interpolate asumirá que es un serie y llenará el campo nulo con el valor siguiente
df.interpolate()
```

```python
#En otra instancia, podemos optar por borrar los valores nulos
df.dropna()
```

```python
#Para eliminar valores nulos de una columna en específico utilizamos la sentencia subset
df.dropna(subset=['Col2'])
```
