```python
import numpy as np
```

```python
arr = np.arange(0,11)
arr
```

##Si no se realiza la función copy, a pesar de ocupar otra variable, nuestro array original se va a modificar

```python
trozo_arr = arr[0:6]
trozo_arr
```

```python
##Cambiamos los valores al trozo de array por cero
trozo_arr[:] = 0
trozo_arr
```

##Si volvemos al arreglo original veremos que tambien ha sido modificado

```python
arr
```

Generemos un nuevo array y utilicemos el comando copy para copiar el arreglo y no modificar el original

```python
arr_dos = np.arange(0,11)
arr_dos
```

```python
arr_copy = arr_dos.copy()
arr_copy[:] = 100
arr_copy
```
