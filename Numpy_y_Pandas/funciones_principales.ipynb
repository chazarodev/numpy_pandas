##Manejo de errores en las dimensiones de los arreglos

```python
c = np.array([[1,2],[3,4]])
d = np.array([5,6])
```

```python
##Error ocurrido debido a que los arreglos no tienen el mismo número de dimensiones
np.concatenate((c,d), axis=0)
```

```python
print(c.ndim)
print(d.ndim)
```

```python
##Expandimos el número de dimensiones para que coincida con el arreglo a y poder concatenar
d = np.expand_dims(d,axis=0)
```

```python
d
```

```python
np.concatenate((c,d), axis=0)
```

```python
##Dado que la lista b ahora tiene una dimensión de dos, concatenarla por el axis 1 ahora produce un error
np.concatenate((c,d), axis=1)
```

```python
##La T mayúscula es la transpuesta de una matriz, es decir, invertir las filas por las columnas y viceversa
np.concatenate((c,d.T), axis=1)
```
