#Realización de operaciones con numpy

```python
##Para poder realizar operaciones, debemos de realizar nnuestros arreglos utilizando numpy
arr = np.arange(1,10)
arr
```

```python
#Creamos una copia para evitar modificar el arreglo original
arr2 = arr.copy()
```

Operaciones

```python
##Ahora la operación matemática si dará cómo resultado los valores esperados
arr_sum = arr2 + 2
arr_rest = arr2 - 2
arr_mult = arr2 * 2
arr_div = arr2 / 2
print(arr_sum)
print(arr_rest)
print(arr_mult)
print(arr_div)
```

```python
#Operaciones entre arrays
arr + arr2
```

Operaciones con matriz

```python
#Generamos una matriz
matriz = arr.reshape(3,3)
matriz
```

```python
##Generamos una copia para evitar modificar la matriz original y hacer operaciones entre esta y la nueva
matriz2 = matriz.copy()
```

```python
matriz + matriz2
```

Operaciones de producto punto

```python
#En ocasiones, debemos transponer una matriz para lograr realizar esta operación
matriz2.T
```

```python
np.matmul(matriz,matriz2.T)
```

```python
#Otra manera de hacerlo es con la siguiente sintaxis
matriz @ matriz2.T
```
