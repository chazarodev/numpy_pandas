```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv')
```

```python
df_books.head(5)
```

#pivot_table

```python
""" 
Como resultado, los valores de Author pasan a formar el índice por fila 
y los valores de Genre pasan a formar parte de los índices por columna, 
mientras que, el User Rating se mantiene como valor. 
"""

df_books.pivot_table(index='Author', columns='Genre', values='User Rating')
```

```python
"""
En esta caso, tenemos por cada género la suma a lo largo de los años, y además de la suma,
podemos obtener más valores como media, desviación estándar, el conteo, la varianza, etc; únicamente,
cambiando el parámetro de la función de agrupamiento aggfunc() 
"""

df_books.pivot_table(index='Genre', columns='Year',values='User Rating', aggfunc='sum')
```

#Melt

El método melt toma las columnas del DataFrame y las pasa a filas, con dos nuevas columnas para especificar la antigua columna y el valor que traía.

```python
df_books[['Name', 'Genre']].head(5)
```

```python
#Aplicando melt cada resultado de las dos columnas pasa a una fila de tipo: llave:valor.
df_books[['Name', 'Genre']].head(5).melt()
```

```python
#De la siguiente manera indico las columnas a las que no quiero hacer melt con el parámetro: id_vars; a las que sí, usando: value_vars
df_books.melt(id_vars='Year',value_vars='Genre')
```

```python
#Con fill_value podemos indicar con que tipo de valor queremos rellenar los datos nulos
df_books.pivot_table(index='Author', columns='Year', values='User Rating', fill_value='-', aggfunc='mean').head(15)
```

```python
#Utilizando los parámetros var_name='' y value_name='' podemos indicar el nomre de nuestras columnas
df_books.melt(id_vars='Year',value_vars='Genre', var_name='Variables', value_name='Valores')
```
