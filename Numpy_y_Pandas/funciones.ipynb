```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',',header=0)
```

```python
df_books.head(2)
```

#Funciones especiales con pandas

```python
#Nos arroja la información de nuestro dataframe
df_books.info()
```

```python
#De las columnas númericas, describe nos arrojará datos estadísticos de estas
df_books.describe()
```

```python
#La función tail es la función opuesta a la función head, pues, mientras Head no retorna los primeros valores de la tabla solicitados, tail no arroja los últimos de la tabla.
df_books.tail(2)
```

```python
df_books.head(2)
```

```python
#Esta función nos indidca cuánta memoria esta ocupando el data frame
df_books.memory_usage(deep=True)
```

```python
#La siguiente función realiza un conteo por el valor que queramos ver la información
df_books['Author'].value_counts()
```

```python
#Si queremos accer a una fila en concreto y visualizar su información podemos realizar lo siguiente:
df_books.iloc[0]
```

```python
#Para duplicar un línea ejecutaremos lo siguiente
df_books = df_books.append(df_books.iloc[0])
df_books
```

```python
#La siguiente función nos ayuda a remover aquellos que se encuentren duplicados.
df_books.drop_duplicates()
```

```python
#Ordenar el dataframe por una columna específica
df_books.sort_values('Year', ascending=False)
```
