```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',',header=0)
```

```python
df_books.head(2)
```

```python
#Observamos que Author ya no es columna sino que se convierte en indice
df_books.groupby('Author').max()
```

```python
#Por tanto, al ser indice podemos acceder a él
df_books.groupby('Author').sum().loc['William Davis']
```

```python
#En caso de no querer resetear el index para que vuelva a columna, realizamos lo siguiente:
df_books.groupby('Author').sum().reset_index()
```

#Funciones de agregación agg()

```python
#Función para obtener el valor mínimo y el valor máximo
df_books.groupby('Author').agg(['min', 'max'])
```

```python
df_books.groupby('Author').agg({'Reviews': ['min', 'max'], 'User Rating': 'sum'})
```

```python
df_books.groupby(['Author', 'Year']).count()
```
