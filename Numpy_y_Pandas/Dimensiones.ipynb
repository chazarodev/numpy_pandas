```python
import numpy as np
```

```python
scalar = np.array(42)
print(scalar)
scalar.ndim
```

```python
vector = np.array([1,2,3])
print(vector)
vector.ndim
```

```python
matriz = np.array([[1,2,3], [4,5,6]])
print(matriz)
matriz.ndim
```

```python
vector = np.array([1,2,3], ndmin = 10)
print(vector)
vector.ndim
```

```python
expand = np.expand_dims(np.array([1,2,3]), axis = 0)
print(expand)
expand.ndim
```

```python
print(vector, vector.ndim)
vector_2 = np.squeeze(vector)
print(vector_2, vector_2.ndim)
```
