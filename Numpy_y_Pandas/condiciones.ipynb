```python
import numpy as np
```

```python
arr = np.linspace(1,10,10, dtype='int8')
arr
```

##Realizar condiciones

Las condiciones utilizando operadores nos retornarán valores booleanos

```python
arr > 5
```

```python
##Si enviamos el código de la siguiente manera, nos filtrará el array por los valores que sean positivos
indices_condicion = arr > 5
arr[indices_condicion]
```

Podemos hacer condiciones multiples, por ejemplo:

```python
arr[(indices_condicion) & (arr < 9)]
```

Podemos cambiar valores a datos que cumplan ciertas condiciones

```python
arr[arr > 5] = 99
arr
```

```python
##Mostrar aquellos que son iguales a 99
arr[arr==99]
```
