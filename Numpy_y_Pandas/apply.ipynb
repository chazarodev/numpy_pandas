```python
import pandas as pd
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv')
df_books.head(2)
```

Creando un función de python

```python
def two_times(value):
  return value * 2
```

```python
#Apply no ayuda a aplicar la función que hemos definido
df_books['Rating_2'] = df_books['User Rating'].apply(two_times)
df_books.head(10)
```

Con apply podemos incluso pasar como parámetro funciones de tipo lambda

```python
df_books['Rating_3'] = df_books['User Rating'].apply(lambda x : x * 3)
df_books.head(10)
```

Aplicando funciones pero a más de una columna

```python
#De esta manera podemos hacer una función aplicada al dataframe en general. Es importante específicar si hacemos referencia a los nombre de fila o columna con axis
df_books['Rating_Fiction'] = df_books.apply(lambda x : x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis=1)
df_books.head(10)
```
