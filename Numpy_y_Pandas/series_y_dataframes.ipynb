```python
import pandas as pd
```

```python
##Creando una lista con pandas
##En el caso de no indicar el index, pandas nos creará una lista iniciando el indice en 0
psg_players = pd.Series(['Navas', 'Mbappe', 'Messi', 'Neymar', 'Donnaruma'],
          index=[1,7,30,10,11]
          )
psg_players
```

```python
#Creación de un diccionario sino es por medio de una lista
dictionary = {1: 'Navas', 7: 'Mbappe', 30: 'Messi', 10: 'Neymar', 11: 'Donnaruma'}
dictionary
```

De esta manera, al panda series podemos enviarle como parametro un diccionario

```python
psg_players_dict = pd.Series(dictionary)
psg_players_dict
```

Al acceder a los objetos es importante tener en cuenta que accedemos por medio de los indices especificados y no partiendo del indice 0 como si fuese una lista array 

```python
psg_players[30]
```

```python
##El código con slicing funciona sin problemas
psg_players[0:3]
```

Se pueden realizar dataframes complejos de la siguiente manera

```python
dictionary_players = {'Jugador': ['Navas', 'Mbappe', 'Messi', 'Neymar', 'Donnaruma'],
 'Altura': [183.0,178.0,177.0,170.0,195.0],
 'Goles': [3,98,562,235,2]}
```

Colocamos la estructura de nuestro diccionario en el parametro de pandas para crear un DataFrame

```python
##De igual manera, si no definimos un index este comenzará contando a partir del 0
df_players = pd.DataFrame(dictionary_players, index=[1,7,30,10,11])
```

```python
df_players
```
