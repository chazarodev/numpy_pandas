```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('/content/bestsellers-with-categories.csv', sep=',',header=0)
```

```python
df_books.head(2)
```

Realizar filtrado por condiciones

```python
##Nos retornará valores booleanos y lo asignaremos a una variable
mayor_2016 = df_books['Year'] > 2016
```

```python
#Ahora podemos llamar al objeto df_books y pasarle como lista los valores que queremos visualizar
df_books[mayor_2016]
```

```python
#Podemos obtener el mismo resultado realizando de la siguiente manera
df_books[df_books['Year'] > 2016]
```

```python
#Si queremos filtrar por una columna que contenga algun texto, ejecutamos la siguiente línea de código
df_books[df_books['Author'].str.contains('Peterson')]
```

```python
#Para evaluar más de una condición, lo lograremos colocando cada una de ellas entre paréntesis
df_books[(df_books['Year'] > 2017) & (df_books['User Rating'] > 4.6) & (df_books['Genre'] == 'Non Fiction') & (df_books['Reviews'] > 15000)]
```

```python
#Lograremos asignar filtros a variables y colocar nuestras condiciones con respecto a estas
year_2010 = df_books['Year'] == 2010
author_book = df_books['Author'] == 'Mark Twain'
#Nos arroja como resultado todos los libros del 2010 escritos por Mark Twain
df_books[year_2010 & author_book]
```

#Negando las condiciones utilizando la virgulilla

```python
"""Si al códgio anterior aplicamos una virgulilla (~) nos arrojará como resultado aquellos que no cumplan esa condición, es decir, todos quellos libros publicados
en el 2010 que no fueron escritos por el autor Mark Twain""" 
df_books[year_2010 & ~author_book]
```
